# String Methods

**The upper(), lower(), isupper(), and islower() Methods**

The upper() and lower() methods are helpful if you need to make a case-insenstive comparison

---
These methods do not change the string itself but return new string values



```python
print('How are you?')
feeling = input()
if feeling.lower() == 'great':
  print('I feel great too.')
else:
  print('I hope the rest of your day is good')
```

If you want to change the original string, you have to call upper() or lower() on the string and then assign the new string to the variable where the original was stored

```python
spam = 'Hello World'
spam = spam.upper()
spam
```

The isupper() and islower() methods will return a Boolean True value if the string has at least one letter and all the letters are uppercase or lowercase. Otherwise it returns a False

```python
spam = 'Hello, world!'
spam.islower()
```

```python
spam.isupper()
```

```python
'HELLO'.isupper()
```

The isX() Methods
---
Helpful when you need to validate user input


isdecimal() --> Returns *True* if the string consists only of numeric characters and is not blank

```python
while True:
  print('Enter your age: ')
  age = input()
  if age.isdecimal():
    break
  print('Please enter a number for your age')
```

isalnum() --> Returns *True* if the string consists only of letters and numbers

```python
while True:
  print('Select a new password (letters and numbers only): ')
  password = input()
  if password.isalnum():
    break
  print('Passwords can only have letters and numbers')
```

startswith() and *endswith()* --> returns True if the string value they are called on begins or ends with the string paased to the method. Otherwise they return *False*

---
Useful aletrnatives to the == equals operator if you need to check only whether the first or last part of the string is equal to another string.


```python
'Hello, world. My name is Pam'.startswith('Hello')
```

```python
'Hello, world. My name is Pam'.endswith('Pam')
```

The *join()* and *split()* Methods

---
*join()* method is ueful when you have a list of strings that need to be joined together into a single string value. It is called on a string value and is passed a list value.

---
*split()* method does the opposite: it's called on a string value and returns a list of strings. By default, string is split wherever whitesapce characters such as space, tab, or newline characters are found.


```python
' '.join(['My', 'name', 'is', 'Pam'])
```

```python
'XXX'.join(['My', 'name', 'is', 'Pam'])
```

```python
'My name is Pam'.split()
```

You can pass a delimiter string to the *split()* method to specify a different string to split on.

```python
'My name is Pam'.split('a')
```

A common use of *split()* is to split a multiline string along the new line character. This lets us split the miltiline string stored in *spam* along the newlines and return a list in which each item correspons to one line of the string.

```python
spam = '''Dear Alice,
Eve's cat has been arrested for catnapping,
cat burglary,
and extortion.
Sincerely,
Bob'''

spam.split('\n')
```

# LAMBDA Functions (Not in textbook)

```python
def add(x, y):
  return x + y

print(add(5, 3))
```

```python
add = lambda x, y:x + y
print(add(5, 3))
```

```python
print((lambda x, y:x + y)(5, 3))
```

```python
(lambda x, y:x + y)(5, 3)
```

Anonymous function inside another function


```python
def makeAdder(n):         # A function that accepts one arg(n), and that arg will be added with unknown number (x)
  return lambda x:x + n         

plus3 = makeAdder(3)    # make a function that always adds the number you send in with 3
plus5 = makeAdder(5)    # # make a function that always adds the number you send in with 5

print(plus3(4))
print(plus5(4))
```
