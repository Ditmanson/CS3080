**Python Operators & Common Data Types**

The following expression will evaluate down to a single, new string value that combines the text of the two strings.

```python
'Alice' + 'Bob'
```

If you try to use the + operator on a string and an integer value, Python will not know how to handle this

```python
'Alice' + 42
```

The * opertator is used for multiplication when it operates on two integer or floating-point values.

```python
10 * 5
```

When the * operator is used on one string value and one integer value, it becomes the string replication operator

```python
'Alice' * 5
```

The following two expressions don't make sense to Python. You can't multiply two words and it's hard to replicate an arbitrary string a fractional number of times.

```python
'Alice' * 'Bob'
```

```python
'Alice' * 5.0
```

---


Evaluate down to a Boolean value: True or False

```python
if (5 < 6) and (6 < 7):
  print('True')
else:
  print('False')

if (1 == 2) or (2 == 2):
  print('True')
else:
  print('False')
```

**if** statements


---

Executes if the statement's condition is True


---

**else** is executed only when if statement's condition is False

```python
print('What is your name?')
name = input()
if name == 'Pam':
  print('Hi, Pam.')
else:
  print('Hello, stranger.')
print('Bye')
```

**elif** statements

```python
print('What is your name?')
name = input()
print('What is your age?')
age = int(input())
if name == 'Alice':
  print('Hi, Alice.')
elif age < 12:
  print('You are not')
elif age > 2000:
  print('That seems a little old to me.')
elif age > 100:
  print('You are not Alice, grannie')
```

**while** Statements


---


The code in a **while** clause will be executed over and over as long as while's statement condition is True.


---
The condition is always checked at the start of each *iteration* (each time the loop is executed.


```python
spam = 0
while spam < 5:
  print('Hello, world.')
  spam = spam + 1
```

At the end of the **while** clause, the program jumps back to the start of the while statement. Compare that to the same code using an **if** statement.

```python
spam = 0
if spam < 5:
  print('Hello, world.')
  spam = spam + 1
```

If the execution in a while reaches a **break** statement, it immediately exists the while loop.

```python
while True:
  print('Please type your name: ')
  name = input()
  if name == 'Pam':
    break

print('Thank you')
```

The following has the same effect

```python
name = ''
while name != 'Pam':
  print('Please type your name: ')
  name = input()

print('Thank you')
```

When the program execution reaches a **continue** statement, the program exection immediately jumps back to the start of the loop and reevalutes the loop's condition. (Similar to what happends when the execution reaches the end of the loop)

```python
while True:
  print('Who are you?')
  name = input()
  if name != 'Pam':
    continue
  print('Hello, Pam. What is the password? (It is a fish.)')
  password = input()
  if password == 'swordfish':
    break
print('Access granted.')
```

**Truthy and Falsey Values**


---
When used in conditions, 0, 0.0 and ' ' (empty string) are considered False.


```python
numOfGuests = int(input())
if numOfGuests:
  print('Be sure to have enough room for all your guests. ')
else:
  print('You do not have any guests.')
```

**for** loop statement


---
First time it runs, variable *i* is set to 0.
After execution, goes back to top of the loop and for statement increments *i* by one.


```python
print('My name is ')
for i in range(5):
  print('Pam Five Times (' + str(i) + ')')
```

```python
print('My name is ')
for i in range(12,16):
  print('Pam Five Times (' + str(i) + ')')
```

```python
print('My name is ')
for i in range(0,10,2):
  print('Pam Five Times (' + str(i) + ')')
```

```python
print('My name is ')
for i in range(5,-1,-1):
  print('Pam Five Times (' + str(i) + ')')
```

Another example of **for** loop.


---

Add up all the numbers from 0 to 100.

```python
total = 0
for num in range(101):
  total = total + num
print(total)
```

You can also use a **while** loop to do the same thing as a **for** loop (**for** loops are more concise)

```python
print('My name is')
i = 0
while i < 5:
  print('Pam Five Times (' +str(i) + ')')
  i = i + 1
```
